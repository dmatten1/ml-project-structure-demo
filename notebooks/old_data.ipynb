```python
raw_data_path = "../data/raw"
csv = [
    f for f in glob(os.path.join(raw_data_path, "*.csv"))
]
curtest_files = [f for f in csv if os.path.basename(f).startswith("curtest")]

curtest_dfs = [pd.read_csv(f) for f in curtest_files]

# Concatenate all DataFrames
curtest_master = pd.concat(curtest_dfs, ignore_index=True)

# Ensure test_id is treated as string (important if it has NaNs or non-strings)
curtest_master['test_id'] = curtest_master['test_id'].astype(str)

# Create mask for unwanted prefixes
mask = curtest_master['test_id'].str.startswith(('X', 'A', 'RD3','W','C','U'))

# Keep only rows that do NOT match the mask
curtest_master = curtest_master[~mask]

curtest_master.to_csv("../data/processed/curtest_master.csv")
```
